# **Importando a Biblioteca e a Base de Dados**

```python
import pandas as pd
```

```python
police_pd = pd.read_csv("police.csv")
display(police_pd)
```

# **ETL da Base de Dados**

```python
police_pd = police_pd.dropna(how='all', axis=1)
display(police_pd)
```

```python
police_pd = police_pd.dropna()
display(police_pd)
```

```python

contagem = police_pd['violation_raw'].value_counts()

police_pd['quantidade'] = police_pd['violation_raw'].map(contagem)

total = len(police_pd)

police_pd['percentual'] = (police_pd['quantidade'] / total) * 100

display(police_pd)
```

```python
police_pd['driver_age_raw'] = police_pd['driver_age_raw'].astype(int)
```

```python
police_pd.to_csv("police_tratado.csv", index=False)
```
